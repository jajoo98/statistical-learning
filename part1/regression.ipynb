```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
def f(x): # define given function
    return x + 3*(x**2) - (1.5)*(x**3)
```

```python
num_sim = 10 # number of regression lines
betas = []
x = np.linspace(-0.5,1.5,1000) # grid to create original curve
y = f(x)
```

```python
plt.plot(x, y, 'b', linewidth=2); # plot the original function
plt.title("Original Function");
plt.axhline(0, color='black')
plt.axvline(0, color='black')
fig = plt.gcf()
fig.set_size_inches(10, 6);
ax = plt.gca()
ax.set_ylim([-1.4, 3.8]);
```

```python
for i in range(num_sim): 
    xtrain = np.random.uniform(-0.5, 1.5, 15) # generate random x points
    ytrain = f(xtrain) # generate y points
    xtrain = sm.add_constant(xtrain)
    model = sm.OLS(ytrain, xtrain) # regression with intercept term
    results = model.fit()
    betas.append(results.params)
```

```python
for i in range(len(betas)): # plot the lines
    plt.plot(x, betas[i][0]+betas[i][1]*x, 'r', linewidth=0.7);
plt.plot(x, y, 'b', label='Original function', linewidth=2);
plt.title("Superimposed Regression Lines");
plt.legend();
plt.axhline(0, color='black')
plt.axvline(0, color='black')
fig = plt.gcf()
fig.set_size_inches(10, 6);
```
